```python
df1['salary'] = df1['salary'].astype(int) 
df1.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,5))
sns.countplot(df1['status'])
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
for i in df1.columns:
    df1[i]=encoder.fit_transform(df1[i])
```

```python
df1.head()
```

```python
x=df1.drop(['status','salary'],axis=1)
y=df1["status"]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=146)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
my_model = LogisticRegression()
result = my_model.fit(x_train, y_train)
```

```python
predictions = result.predict(x_test)
predictions
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
confusion_mat = confusion_matrix(y_test, predictions)
confusion_df = pd.DataFrame(confusion_mat, index=['Actual negative','Actual positive'], columns=['Predicted negative','Predicted positive'])
confusion_df
```

```python
Color_conf_matrix = sns.heatmap(confusion_df, cmap='coolwarm',annot = True)
```
